```python
import ibm_db_dbi as db2
conn_read = db2.connect()

import pandas as pd
dataset1 = pd.read_sql("SELECT * FROM JUPYTER_04.VEICOLI", conn_read)
```

```python
print(dataset1)
```

```python
#esempio di filtraggio su colonna
x = dataset1[dataset1['NUMERO_VEICOLI'] > 2500]
```

```python
#estrae la prima variabile (anno) di tutti i casi con TIPOLOGIA_VEICOLO = 'AUTOBUS' 
x = dataset1[dataset1['TIPOLOGIA_VEICOLO'] == 'AUTOBUS'].iloc[:,0]
```

```python
print(x)
```

```python
#estrae la terza variabile (numero_veicoli) di tutti i casi con TIPOLOGIA_VEICOLO = 'AUTOBUS' 
y = dataset1[dataset1['TIPOLOGIA_VEICOLO'] == 'AUTOBUS'].iloc[:,2]
```

```python
print(y)
```

```python
#disegna grafico a linee
plt.plot(x, y)
plt.title("TITLE", color = "blue")
plt.xlabel("Anno acquisto", color = "purple")
plt.ylabel("Numero veicoli", color = "green")
plt.grid(True)
plt.legend(['Autobus a Milano'])
```

```python
#disegna grafico con rappresentazione con istogramma e salvataggio immagine con 300 dpi
plt.bar(x, y)
plt.title("TITLE", color = "blue")
plt.xlabel("Anno acquisto", color = "purple")
plt.ylabel("Numero veicoli", color = "green")
plt.grid(True)
plt.legend(['Autobus a Milano'])
plt.savefig("test2.png", dpi = 300)
```
